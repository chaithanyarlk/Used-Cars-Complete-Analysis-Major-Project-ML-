```python
def detect_outliers(data):
    outliers = []
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)
    for i in data :
        z_score = (i-mean)/std
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return (outliers)
```

```python
print(detect_outliers(df_drop['Kilometers_Driven']))
print(detect_outliers(df_drop['Seats']))
print(detect_outliers(df_drop['Price']))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_param = {
    'knn':{
        'model':KNeighborsClassifier(),
        'params':{
            'n_neighbors':range(1,5,1),
        },
    },
    'svm':{
        'model': svm.SVC(gamma = 'auto'),
        'params':{
            'C':[1,10,20,30],
            'kernel':['rbf','linear']
        },
    },
    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[1,5,10,20,30]
        }
    },
    'logistic_regression':{
        'model':LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{
            'C':[1,5,10,20]
        }
    }
    
}
```

```python
scores=[]
for model_name ,mp in model_param.items():
    print(model_name)
    clf = GridSearchCV(mp['model'],mp['params'],cv=3,return_train_score=False)
    print('done with GridSearchCV')
    clf.fit(X_train,y_train)
    print('done with fitting')
    scores.append({
        'model':mp,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
    print('done with appending score list')
    y_predict = clf.predict(X_test)
    print('done with predicting')
    print(accuracy_score(y_test,y_predict))
    print(mean_absolute_error(y_test,y_predict))
    print(mean_squared_error(y_test,y_predict))
    print(r2_score(y_test,y_predict))
```
